```python
import easyocr

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
reader = easyocr.Reader(['ne','en'], gpu=False) 
```

```python
img_path = '../data/img3.jpg'

image = mpimg.imread(img_path)  
plt.imshow(image)
plt.axis("off") 
plt.show()
```

```python
result = reader.readtext(img_path, detail = 0)
```
